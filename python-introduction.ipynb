## Conditions and loops

**Indentation**: In python, to consider a code to be part of a loop or a if statement, the code must be indented with [TAB] under the loop, if, or function (seen later) definition.

````python
"""
if statement
An `if statement` is used to execute a block of code only if a condition is True.

The block of code is indented.
  if (condition is True):
    (code to be executed)
  else:
    (code to be executed)
"""
if 1 == 1:
  print("One is equal to one")
if 1 != 1:
  print("One is not equal to one")
else:
  print("Is one equal to one?")

# Note: 1)  since the statement 1 != 1 is False, the indented code following
#           "if 1 != 1:" wil not be executed.
#       2)  The first if statement is process independently from the next
#           if-else code block.
````

````python
"""
else-if statement
An `else-if statement` is used to execute a block of code only if the previous condition is False.
  if (condition_1 is True):
    (code to be executed)
  elif (condition 1 is False and condition_2 is True):
    (code to be executed)
  else:
    (code to be executed)
"""
print("-Exemple 1-")
if 1 == 1:
  print("One is equal to one")
elif 1 != 1:
  print("One is not equal to one")
else:
  print("Is one equal to one?")
print("\n")

print("-Exemple 2-")
if 1 != 1:
  print("One is not equal to one")
elif 1 == 1:
  print("One is equal to one")
else:
  print("Is one equal to one?")

# Note: Adding the expression elif allows the use of more than one if statement
#       in the same if-else code block.
````

````python
"""
Combined conditions in if statement
  and: condition 1 is True and condition 2 is True
  or: condition 1 is True or condition 2 is True
"""
if 1==1 and 2!=1:
  print("With 'and': all the conditions are True")
if 1==1 or 2==1:
  print("With 'or': one of the conditions have to be True")
````

***Loops*** are operations that run a code block over and over for a certain
predefine *number of iteration* **or** until a *condition* is met.


````python
"""
The "For" iteration loop
  for (element) in (elements):
    (code to execute)
"""
neuron_structures = ["Dendrite", "Soma", "Nucleus", "Axon"]

for structure in neuron_structures:
  print("Code block start~")
  print(structure)
  print("Code block end~\n")

# Note: The variable structure of this exemple takes the value of
#       the list elements one by one through loop iterations.

"""
It is possible to `break` out of a loop or `continue` to the next iteration when a condition is met.
The `continue` statement will skip the rest of the code block and go to the next iteration.
  for (element) in (elements):
    if (condition):
      break
    elif (condition):
      continue
    (code to execute)
"""
for structure in neuron_structures:
  if structure == "Nucleus":
    break
  elif structure == "Soma":
    continue
  print(structure)
````

````python
"""
Sometimes, it is useful to have the index of the element in the list as well. This can 
be done using the `enumerate` function. The `enumerate` function is equivalent to:
  index = 0
  for element in elements:
    (code to execute)
    index += 1
  
The "For" iteration loop using `enumerate`
  for (index, element) in enumerate(elements):
    (code to execute)
"""
neuron_structures = ["Dendrite", "Soma", "Nucleus", "Axon"]

for index, structure in enumerate(neuron_structures):
  print(structure, "is at index", index, "of neuron_structures list.")

# Note: enumerate returns the positional index of the element and the element
#       itself.
````

````python
"""
The "While" conditional loop
  while (condition):
    (code to execute)
"""
# The while loop must have a stop criterion, leading, otherwise to an infinite
# loop.
finish_criterion = False
iteration = 0
while finish_criterion == False:
  print("iteration: ", iteration)
  iteration += 1
  if iteration > 5:
    finish_criterion = True # Assign True to variable
````

````python
"""
The `for` loop can be used to iterate over a range of numbers.
"""
for i in range(5):
    print(i)

"""
The `items`, `keys`, and `values` function can be used to iterate over a dictionary.
"""
xp_datas = {
    "Neuron_IDs": [1, 2, 3],
    "Neuron_structures": ["Dendrite", "Soma", "Nucleus", "Axon"],
    "Experiment": "14_KO_338"
}
for key, value in xp_datas.items():
    print(key, value)

for key in xp_datas.keys():
    print(key)

for value in xp_datas.values():
    print(value)
````

## Functions

A ***function*** in Python is a *block of organized, reusable code* that is used to perform a single, related action. This can perform the action when the function is called with or without returning values.

It is defined using the `def` keyword followed by the function name and parentheses `()`. The code block within every function starts with a colon `:` and is indented with [TAB].

It is highly recommended to use function names that are descriptive of what the function does. The same goes for the arguments of the function. For example, if a function calculates an exponential decay function, it is better to name it `exp_decay` than `f`.

````python
"""
Functions
  def function_name(argument1, argument2,...):
    (code to execute)
    return (values to return)
Use:
  value = function_name(argument1 = var1, argument2 = var2, ...)
"""
def multiply_integer(integer1, integer2):
  result = integer1 * integer2
  return result

value1 = 4
value2 = 5

value3 = multiply_integer(integer1=value1, integer2=value2)
print(value3)

# Note: It could be argued that naming the function multiply_integer is redundant since the function could also be used to multiply floats.
````

# Advanced Topics

## Object-Oriented Programming

Object-oriented programming (OOP) is a programming paradigm that uses objects and classes to design and build applications. It focuses on creating reusable code and organizing it into classes that represent real-world entities. OOP concepts include inheritance, encapsulation, polymorphism, and abstraction, which help developers create modular, maintainable, and scalable code.

In Python, everything is an object, and classes are used to define new types of objects. Classes can have attributes (variables) and methods (functions) that define their behavior. Objects are instances of classes, and they can interact with each other through methods and attributes.

## Classes and Objects

In Python, a class is defined using the `class` keyword followed by the class name. The class body contains attributes and methods that define the behavior of the class. To create an object of a class, we call the class name as if it were a function, which creates a new instance of the class.

```python
class MyClass:
    def __init__(self, attribute1, attribute2):
        self.attribute1 = attribute1
        self.attribute2 = attribute2

    def my_method(self):
        return self.attribute1 + self.attribute2
```

## Inheritance

In Python, inheritance allows a class to inherit attributes and methods from another class. This enables code reuse and promotes a hierarchical structure in object-oriented programming. The class that inherits from another class is called a subclass or derived class, while the class being inherited from is called a superclass or base class.

```python
class ParentClass:
    def __init__(self, attribute1):
        self.attribute1 = attribute1

    def parent_method(self):
        return self.attribute1

class ChildClass(ParentClass):
    def __init__(self, attribute1, attribute2):
        super().__init__(attribute1)
        self.attribute2 = attribute2

    def child_method(self):
        return self.attribute1 + self.attribute2
```

## Typing

Python is a dynamically typed language, meaning that variable types are determined at runtime. This flexibility allows for rapid development and prototyping but can lead to errors if not handled carefully. To address this issue, Python introduced type hints in PEP 484, which allows developers to specify types for variables, function arguments, and return values.
